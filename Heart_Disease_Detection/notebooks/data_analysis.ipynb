# Heart Disease Data Analysis

This notebook covers preprocessing, summary statistics, and exploratory visualizations for the heart disease dataset.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve().parents[0]
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

import matplotlib.pyplot as plt
import seaborn as sns
from utils.data_processing import create_clean_dataset, dataset_summary

sns.set_theme(style='whitegrid')
```

```python
df = create_clean_dataset()
df.head()
```

```python
dataset_summary(df)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
sns.histplot(df['age'], kde=True, ax=axes[0, 0], color='steelblue')
sns.histplot(df['chol'], kde=True, ax=axes[0, 1], color='tomato')
sns.boxplot(data=df[['trestbps', 'chol', 'thalach']], ax=axes[1, 0])
sns.countplot(x='target', data=df, ax=axes[1, 1], palette='Set2')
axes[0, 0].set_title('Age Distribution')
axes[0, 1].set_title('Cholesterol Distribution')
axes[1, 0].set_title('Boxplots for Key Numerical Features')
axes[1, 1].set_title('Target Class Distribution')
plt.tight_layout()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
